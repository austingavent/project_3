```python
from pathlib import Path
import pandas as pd
import os
import csv
import numpy as np
import requests
import time
import io
import yfinance as yf
import warnings
%load_ext jupyternotify


root_dir = 'Desktop/data_dump/'
len(os.listdir(root_dir))
```

```python

#how to find dates between 1/1/2020 and 12/31/2021
#initalize date window
start_date = '2016-01-01'
end_date = '2020-12-31'

etf_list = ['MTUM', 'SPY', 'NDAQ', 'VOO' ]
materials_list = ['VALE', 'RIO', 'BHP']
media_list = ['TWTR', 'NFLX', 'DIS']
tech_list = ['MSFT', 'TSLA', 'GOOG', 'AAPL', 'FB']
retail_list = ['BABA', 'AMZN', 'WMT' , 'BBY']
finance_list = ['GS', 'JPM', 'BOA']
military_list = ['BA', 'LMT', 'RTX']
food_list = ['PG', 'SBUX', 'CPG', ]
symbols_list = etf_list + materials_list + media_list + tech_list + retail_list + finance_list + military_list + food_list  
len(symbols_list)
```

```python
def get_data(symbol, start_date, end_date):
#Create master list dictionary, to store the each symbols data frame
     
    df_list=[]
    master_list={}
    for each_symbol in symbol: 
        #master_list[each_symbol]=[]
 
    # for each_file in range(len(os.listdir(root_dir))):#test code with first 2 entries to reduce time
    for each_file in os.listdir(root_dir)[:5]:
      if each_file.split('.')[-1]=='csv': #grabbing all files that end with csv
          print(f'Opening {each_file}...')#prints when its reading each file
          with open(root_dir+'/'+each_file) as f: 
              csvreader=csv.reader(f)
              next(csvreader)
              for each_line in csvreader: 
                try: 
                  for each_symbol in symbol: 
                    if each_line[1]==each_symbol and each_line[13]>start_date and each_line[13]<end_date: #Symbol is the 2nd index or in python index 1
                        master_list[each_symbol].append(each_line) 
                except: 
                  pass                                                       
    for each_symbol in symbol: 
      df=pd.DataFrame(master_list[each_symbol], columns=['optionkey', 'Symbol', 'ExpirationDate', 'AskPrice', 'AskSize', 'BidPrice', 'BidSize', 'LastPrice', 'PutCall', 'StrikePrice', 'Volume', 'openinterest', 'UnderlyingPrice', 'DataDate'])
      df_list.append(df)
    # print(master_list)
    return df_list
```
