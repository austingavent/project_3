```python
import pandas as pd
import numpy as np
df = pd.read_csv('sectorized.csv')
df
```

```python
#Organize by sector

etf_list = ['MTUM', 'SPY', 'NDAQ', 'VOO' ]
materials_list = ['VALE', 'RIO', 'BHP']
media_list = ['TWTR', 'NFLX', 'DIS']
tech_list = ['GOOG', 'AAPL', 'FB', 'MSFT', 'TSLA'] #missing msft and tsla
retail_list = ['BABA', 'AMZN', 'WMT' , 'BBY']
finance_list = ['GS', 'JPM']
military_list = ['BA', 'LMT', 'RTX']
food_list = ['PG', 'SBUX', 'CPG' ]

sectors = [etf_list, materials_list, media_list, tech_list, finance_list, military_list, food_list]



def classify_sector(df):
    if df['symbol'].isin(materials_list):
        return 'materials'
    if df['symbol'].isin(media_list):
        return 'media'
    if df['symbol'].isin(tech_list):
        return 'tech'
    if df['symbol'].isin(food_list):
        return 'food'
    if df['symbol'].isin(finance_list):
        return 'finance'
    if df['symbol'].isin(military_list):
        return 'military'
    if df['symbol'].isin(etf_list):
        return 'etf'
```

```python
df['sector'] = df.apply(classify_sector, axis = 1)
df
```
