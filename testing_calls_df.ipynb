```python
def selling_calls_blp(og):
    df = og.loc[(og['bollinger_signal']== -1) & (og['Put signals'] == 1)]
    df['log'] = df['AskPrice'].astype(float) - df['xp'].astype(float)
   # df = df[]
    #df_total = round(df['log'].sum())
    
    df_otm_xs = df.loc[df['otm_xs'] == True]
    df_otm_xs = df_otm_xs[['log']]
    df_otm_xs.to_csv(f'azplease{symbol}selling_calls_blp_otm_xs.csv')
    
    df_otm_xm = df.loc[df['otm_xm'] == True]
    #df_otm_xm['profit'] = round(df_otm_xm['profit'])
    df_otm_xm = df_otm_xm[['log']]
   # df_otm_xm.to_csv(f'apleasebearish_{symbol}selling_calls_blp_otm_xm.csv')

    df_otm_xf = df.loc[df['otm_xf'] == True]
    df_otm_xf['profit'] = round(df_otm_xf['profit'])
    df_otm_xf = df_otm_xf['profit']
    #df_otm_xf.to_csv(f'bearish_{symbol}selling_calls_blp_otm_xf.csv')
    
    df_ctm_xs = df.loc[df['ctm_xs'] == True]
    df_ctm_xs['profit'] = round(df_ctm_xs['profit'])
    df_ctm_xs = df_ctm_xs['profit']
    #df_ctm_xs.to_csv(f'bearish_{symbol}selling_calls_blp_ctm_xs.csv')
    
    df_ctm_xf = df.loc[df['ctm_xf'] == True]
    df_ctm_xf['profit'] = round(df_ctm_xf['profit'])
    df_ctm_xf = df_ctm_xf['profit']
    #df_ctm_xf.to_csv(f'bearish_{symbol}selling_calls_blp_ctm_xf.csv')
    
    df_ctm_xm = df.loc[df['ctm_xm'] == True]
    df_ctm_xm['profit'] = round(df_ctm_xm['profit'])
    df_ctm_xm = df_ctm_xm['profit']
    #df_ctm_xm.to_csv(f'bearish_{symbol}selling_calls_blp_ctm_xm.csv')
    #df_ctm_xf['profit'] = df_ctm_xf['profit']
    
    #df_ctm_xf['profit'] = round(df_ctm_xf['profit'])
    #df_ctm_xf_total = df_ctm_xf[:, 0].sum()

    #print(f'blindly selling calls on bollinger and put signals would net:{df_total}')
    #print(f'buying close to the money, expiring far calls on bollinger and call signals would net: {df_ctm_xf_total}' )
    
atm = .07
ctm = .1
otm = .15
ftm = .2
xf = 28
xm = 19
xs = 7


def get_leaps(og):

    atm_leaps = og.loc[(og['time delta'] > np.timedelta64(300, 'D')) & (og['% out'] > -.01) & (og['% out'] < atm)]
    atm_leaps ['profit'] = round(atm_leaps['profit'] * -1)
    atm_leaps = atm_leaps['profit']
    #atm_leaps.to_csv(f'bullish{symbol}_atm_leaps.csv')

    ctm_leaps = og.loc[(og['time delta'] > np.timedelta64(300, 'D')) & (og['% out'] > atm) & (og['% out'] < ctm)]
    ctm_leaps ['profit'] = round(ctm_leaps['profit'] * -1)
    ctm_leaps = ctm_leaps['profit']
    #ctm_leaps.to_csv(f'bullish{symbol}_ctm_leaps.csv')

    otm_leaps = og.loc[(og['time delta'] > np.timedelta64(300, 'D')) & (og['% out'] > ctm) & (og['% out'] < otm)]
    otm_leaps ['profit'] = round(otm_leaps['profit'] * -1)
    otm_leaps = otm_leaps['profit']
    #otm_leaps.to_csv(f'bullish{symbol}_otm_leaps.csv')

    ftm_leaps = og.loc[(og['time delta'] > np.timedelta64(300, 'D')) & (og['% out'] > otm) & (og['% out'] < ftm)]
    ftm_leaps ['profit'] = round(ftm_leaps['profit'] * -1)
    ftm_leaps = ftm_leaps['profit']
    #ftm_leaps.to_csv(f'bullish{symbol}_ftm_leaps.csv')

def buying_calls_df(og):
    df = og.loc[(og['bollinger_signal']== 1) & (og['Call signals'] == 1)]
    df['log'] = df['AskPrice'].astype(float) - df['xp'].astype(float)
    #df_total = round(df['sclog'].sum() * -1)
##
    df_otm_xs = df.loc[df['otm_xs'] == True]
    df_otm_xs['log'] = (df_otm_xs['log'] * -1)
    df_otm_xs = df_otm_xs[['log']]

    
    
    #return df_otm_xc
    df_otm_xs.to_csv(f'azbullish_{symbol}buying_calls_df_otm_xs.csv')
    
    df_otm_xm = df.loc[df['otm_xm'] == True]
    df_otm_xm['log'] = round(df_otm_xm['log'] * -1)
    df_otm_xm = df_otm_xm[['log']]
    #df_otm_xm.to_csv(f'bullish_{symbol}buying_calls_df_otm_xm.csv')

    df_otm_xf = df.loc[df['otm_xf'] == True]
    df_otm_xf['log'] = round(df_otm_xf['log'] * -1)
    df_otm_xf = df_otm_xf[['log']]
    #df_otm_xf.to_csv(f'bullish_{symbol}buying_calls_df_otm_xf.csv')
    
    df_ctm_xs = df.loc[df['ctm_xs'] == True]
    df_ctm_xs['log'] = round(df_ctm_xs['log'] * -1)
    df_ctm_xs = df_ctm_xs[['log']]
    #df_ctm_xs.to_csv(f'bullish_{symbol}buying_calls_df_ctm_xs.csv')
    
    df_ctm_xf = df.loc[df['ctm_xf'] == True]
    df_ctm_xf['log'] = round(df_ctm_xf['log'] * -1)
    df_ctm_xf = df_ctm_xf[['log']]
    #df_ctm_xf.to_csv(f'bullish_{symbol}buying_calls_df_ctm_xf.csv')
    
    df_ctm_xm = df.loc[df['ctm_xm'] == True]
    df_ctm_xm['log'] = round(df_ctm_xm['log'] * -1)
    df_ctm_xm = df_ctm_xm[['log']]
    #df_ctm_xm.to_csv(f'bullish_{symbol}buying_calls_df_ctm_xm.csv')
    #df_ctm_xf['profit'] = df_ctm_xf['profit']
    
    #df_ctm_xf['profit'] = round(df_ctm_xf['profit'])
    #df_ctm_xf_total = df_ctm_xf[:, 0].sum()

   # print(f'blindly buying calls on bollinger and call signals would net:{df_total}')
    #print(f'buying close to the money, expiring far calls on bollinger and call signals would net: {df_ctm_xf_total}' )

def test_calls(og):
    buying_calls_df(og)
    get_leaps(og)
    selling_calls_blp(og)
```
